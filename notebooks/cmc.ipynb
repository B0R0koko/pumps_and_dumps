```python
from pathlib import Path

import pandas as pd
import os


ROOT_DIR = Path(os.getcwd()).parent
```

```python
df = pd.read_csv(os.path.join(ROOT_DIR, "data/cmc/cmc_snapshots.csv"))

top_100 = df[df["cmcRank"] >= 100]["symbol"].tolist()
```

```python
cmc_top_100 = set([
    el+"BTC" for el in top_100
])
```

```python
data_collected = set(
    os.listdir(os.path.join(ROOT_DIR, "data/trades"))
)
```

```python
target_tickers = list(data_collected.intersection(cmc_top_100))
```

```python
for ticker in os.listdir(os.path.join(ROOT_DIR, "data/trades")):
    if ticker not in target_tickers:
        print(ticker)
```
